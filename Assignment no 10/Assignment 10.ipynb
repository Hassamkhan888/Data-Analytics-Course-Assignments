## ***19. Who Has Better Chance To Get Salary >50k Male or Female***

```python
gender_total = df.groupby("gender")["income"].count()
gender_high_income = df[df['income'] == 1].groupby("gender")["income"].count()

gender_income_proportion = (gender_high_income / gender_total) * 100
print(gender_income_proportion)

better_chance_gender = "Male" if gender_income_proportion["Male"] > gender_income_proportion["Female"] else "Female"
print(f"\n{better_chance_gender} have a better chance of earning >50K.")
```
